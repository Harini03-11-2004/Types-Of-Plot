```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
np.random.seed(25)
data = {
    "x" : np.arange(1,9),
    "y" : np.random.randint(15,50,size=8)
}
df = pd.DataFrame(data)
print(df)
plt.figure(figsize=(6,6))
plt.plot(df["x"],df["y"],marker='*',color = 'purple')
plt.xlabel("x")
plt.ylabel("y")
plt.title("sales Analysis")
plt.grid(True)
plt.show()
```
