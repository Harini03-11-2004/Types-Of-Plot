```python
#write a python program to plot a basic line graph for the function for the values of x from 0 to 10 using matplotlib
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 11)
y = x  

plt.plot(x, y, marker='o',color='red')
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Basic Line Graph")
plt.grid(True)
plt.show()
```

```python
#create a vertical bar chart for the products A,B,C,D with corresponding sales values 40,70,30 and 90
import matplotlib.pyplot as plt
products = ["A", "B", "C", "D"]
sales = [40, 70, 30, 90]
plt.bar(products, sales, color="skyblue")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Sales of Products")
plt.grid(True)
plt.show()
```

```python
#create a 2x2 grid of subplots containing any 4 types of plots
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 30, 25]
data = y  
plt.figure(figsize=(5,5))
plt.subplot(2, 2, 1)
plt.scatter(x, y, color='purple')
plt.title("Scatter Plot")
plt.grid(True)
plt.subplot(2, 2, 2)
plt.bar(x, y, color='purple')
plt.title("Bar Plot")
plt.grid(True)
plt.subplot(2, 2, 3)
plt.barh(x, y, color='purple')
plt.title("Horizontal Bar Plot")
plt.grid(True)
plt.subplot(2, 2, 4)
plt.hist(data, bins=10, color='skyblue', edgecolor='black')
plt.title("Histogram Plot")
plt.grid(True)

plt.grid()
plt.show()

```

```python
#generate 200 random values from a normal distribution and plot a histogram with 20 bins and black edge around the bars
import matplotlib.pyplot as plt
import numpy as np
data = np.random.randn(200)  
plt.hist(data, bins=20, edgecolor='black')
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.grid(True)
plt.show()
```
